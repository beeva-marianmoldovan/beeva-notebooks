```python
def clean(x):
   x = unicodedata.normalize('NFKD', x).encode('ascii','ignore').lower()
   replace_punctuation = string.maketrans(string.punctuation, ' '*len(string.punctuation))
   x = x.translate(replace_punctuation)
   x = re.sub('@%$&[\n/:!,;)()_?¿¡<>]', ' ', x)
   x = re.sub(' - ', ' ', x)
   x = re.sub(' +',' ', x).strip()
   return x

documentos = []
info = []
def creaConjuntoDocumentos(pathLibros):
    for libroFile in os.listdir(pathLibros):
        #print(pathLibros + libroFile)
        lineasDocumento = []
        fileName = pathLibros + libroFile
        if fileName.endswith(".keywords"):
            with open(fileName, 'r') as fileData:
                for lineas in fileData:
                    line = lineas.decode('utf-8')
                    line = clean(line) # ¿? problemas con gensim y tildes y eñes...
                    if len(line) > 1:
                        lineasDocumento.append(line)
                info.append(lineasDocumento)    
                fileData.close()
        if fileName.endswith(".txt"):
            with open(fileName, 'r') as fileData:
                # print "Leo datos, detecto idioma y preproceso."
                for lineas in fileData:
                    idioma = detect(lineas.decode('utf-8'))
                    if idioma == u'es':
                    #if True:
                        #Formatear linea si hace falta aquí
                        lineArray = lineas.split(".")
                        for line in lineArray:
                            if len(line) > 1:
                                line = line.decode('utf-8')
                                line = clean(line) # ¿? problemas con gensim y tildes y eñes...
                                if len(line) > 1:
                                    lineasDocumento.append(line)

                if len(lineasDocumento) > 0:
                    stringDocumento = ' '.join(lineasDocumento)
                    documentos.append(stringDocumento)

                fileData.close()
```

### Comparing similarities LSI

We want to determine similarity between pairs of documents, or the similarity between a specific document and a set of other documents (such as a user query vs. indexed documents). Unlike modern search engines, here we only concentrate on a single aspect of possible similarities—on apparent semantic relatedness of their texts (words). No hyperlinks, no random-walk static ranks, just a semantic extension over the boolean keyword match

```python
fileName = '/home/nievesabalos/Escritorio/nlp/datasets/TED/data txt/1.txt'
lineasDocumento = []
with open(fileName, 'r') as fileData:
    print "Leo datos y preproceso.\n"
    for lineas in fileData:
        #Formatear linea si hace falta aquí
        lineArray = lineas.split(".")
        for line in lineArray:
            if len(line) > 1:
                line = line.decode('utf-8')
                line = clean(line) # ¿? problemas con gensim y tildes y eñes...
                if len(line) > 1:
                    lineasDocumento.append(line)
    # un fichero, un resumen, un documento:
    stringDocumento = ' '.join(lineasDocumento)
    documentos.append(stringDocumento)
    #print documentos
    fileData.close()

print lineasDocumento
```
