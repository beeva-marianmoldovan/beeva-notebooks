```python
def sacar_cp_url(elems_calle, indice):
    calle_url = ''
    for gram in xrange(indice,len(elems_calle)):
        elem = elems_calle[gram].replace(",","")
        if(gram==len(elems_calle)-1):
            calle_url += elem
        else:
            calle_url += elem + "%20"
    return calle_url
```

```python
def saca_calle(elems_calle):
    datos = []
    nombredir = ''
    calle = ''
    indice = 0
    cp = ''
    for gram in elems_calle:
        gram = gram.replace(",","")
        #print gram
        try:
            int(gram)
            cp = sacar_cp_url(elems_calle, indice)
            datos.append(nombredir)
            datos.append(calle)
            datos.append(cp)
            return datos
        
        except:
            nombredir += gram
            calle += gram + "%20"
            indice = indice+1
```

# Análisis de las imágenes de la calle

#Extracción de características con deeplearning

```python
images = gl.image_analysis.load_images(dircalle, recursive=True, random_order=False, with_path=True)
```

#Target en SFrame

```python
dataset_street = gl.SFrame()
dataset_street['image'] = gl.image_analysis.resize(images['image'], 256, 256, 3)
dataset_street['path'] = images['path']
dataset_street['location'] = dataset_ori[edificio_target]['location']
dataset_street['ori'] = 'ref'
```

```python
pretrained_model = gl.load_model('http://s3.amazonaws.com/GraphLab-Datasets/deeplearning/imagenet_model_iter45')
```

```python
dataset_street['extracted_features'] = pretrained_model.extract_features(dataset_street)
```

```python
dataset_street = dataset_street.append(dataset_ori[edificio_target:edificio_target+1])
```

```python
dataset_street = dataset_street.add_row_number()
```

```python
dataset_street.groupby(["ori"],
     {"elems": gl.aggregate.CONCAT("id")})
```

```python
target_id = dataset_street.num_rows()-1
```

```python
target_id
```

#Calcular vecinos más próximos

```python
nearest_neighbor_model = gl.nearest_neighbors.create(dataset_street, features=['extracted_features'], distance='manhattan')
```

```python
similar_images = nearest_neighbor_model.query(dataset_street, k = 3)
```

```python
similar_images = similar_images[similar_images['distance'] > 0]
```

```python
similar_images
```

```python
sg_similarities = gl.SGraph().add_edges(similar_images, src_field='query_label', dst_field='reference_label')
```

```python
sp = gl.shortest_path.create(sg_similarities, weight_field='distance', source_vid=target_id)
sp_graph = sp['graph'] 
```

```python
subgraph = sp_graph.get_neighborhood(ids=[target_id], radius=1, full_subgraph=True)
cluster = subgraph.get_vertices()
```

```python
cluster = cluster[cluster['distance'] > 0.0]
print cluster
```
